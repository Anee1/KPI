```python
data['Date_premier_contact'] = pd.to_datetime(data['Date_premier_contact'])
data['Date_conversion'] = pd.to_datetime(data['Date_conversion'])
```

```python
# Date de référence = aujourd’hui
date_ref = pd.Timestamp.today()

# Fenêtre 3 mois glissants
date_3_mois = date_ref - pd.DateOffset(months=3)

df_3_mois = data[
    (data['Statut'] == 'Converti') &
    (data['Date_conversion'] >= date_3_mois)
]
```

```python
#Taux de conversion global

taux_conversion = (
    data[data['Statut'] == 'Converti'].shape[0] /
    data.shape[0]
)

print(f"Taux de conversion : {taux_conversion:.2%}")
```

```python
#Délai médian de conversion
df_conversion = data[data['Statut'] == 'Converti'].copy()

df_conversion['Delai_conversion'] = (
    df_conversion['Date_conversion'] - df_conversion['Date_premier_contact']
).dt.days

delai_median = df_conversion['Delai_conversion'].median()

print(f"Délai médian de conversion : {delai_median} jours")
```

```python
#Montant souscrit total

montant_total = (
    data[data['Statut'] == 'Converti']['Montant_souscription']
    .sum()
)

print(f"Montant souscrit total : {montant_total:,.0f}")
```

```python
pipeline_actif = data[data['Statut'] != 'Converti']

nb_pipeline = pipeline_actif['Prospect_Nom'].nunique()

print(f"Pipeline actif non converti : {nb_pipeline} prospects")
```

```python
pipeline_actif['Age_pipeline'] = (
    pd.Timestamp.today() - pipeline_actif['Date_premier_contact']
).dt.days

pipeline_ancien = pipeline_actif[pipeline_actif['Age_pipeline'] > 150]

print(f"Pipeline dormant (>30j) : {pipeline_ancien.shape[0]} prospects")
```
